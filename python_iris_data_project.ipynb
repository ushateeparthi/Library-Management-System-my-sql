```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
```

```python
iris_df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
# Display basic information about the dataset
print(iris_df.info())

```

```python
# Summary statistics of numerical columns
print(iris_df.describe())
```

```python
# Pairplot to visualize relationships between variables
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
# Boxplot to visualize distributions
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop('species', axis=1))
plt.title('Boxplot of Iris Features')
plt.show()
```

```python
# Calculate the mean of each feature for each species
means = iris_df.groupby('species').mean()
print(means)

```

```python
# Perform ANOVA to test if there are significant differences between species
from scipy.stats import f_oneway

setosa = iris_df[iris_df['species'] == 'setosa']['petal_length']
versicolor = iris_df[iris_df['species'] == 'versicolor']['petal_length']
virginica = iris_df[iris_df['species'] == 'virginica']['petal_length']

f_statistic, p_value = f_oneway(setosa, versicolor, virginica)
print("F-Statistic:", f_statistic)
print("P-value:", p_value)
```
